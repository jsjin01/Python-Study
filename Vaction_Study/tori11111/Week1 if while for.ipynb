for 문
for 변수 in 시퀀스(리스트, 튜플, 문자열 등) : 시퀀스를 순회하며 반복 실행
for 변수 in range(start, stop, step) : start 값만 설정 시 : 0부터 start까지 stop 까지 설정 시 : start부터 stop - 1까지 step 설정 시 : 증가 값이 설정 됨

함수 안에서 선언된 변수의 효력 범위

*args 를 사용하여 여러 개의 인수를 한번에 받을 수 있음(list 처럼 사용 가능)
예시)
def sum_many(*args):
    sum = 0
    for i in args:
        sum = sum + i
    return sum

**kwargs를 사용하여 여러 개의 키워드 인수를 받을 수 있음(Dictionary 처럼 사용 가능)
예시)
def print_kwargs(**kwargs):

전역변수와 지역 변수 
전역 변수(Global Variable): 함수 외부에서 선언되어 프로그램 전체에서 접근 가능. 
지역 변수(Local Variable): 함수 내부에서 선언되어 해당 함수에서만 사용 가능.
                        ㄴ 별도의 변수여서 프로그램에 영향을 끼치지 않는다.
예시
a = 1
def vartest(a):
    a = + 1

vartest(a)
print(a)
<결과>
a = 1

함수 내부에서 전역 변수를 변경하기 위해서는 global 키워드를 선언해야 됨

예시 1)
a = 1
def vartest():
    global a
    a = + 1

vartest(a)
print(a)
<결과>
a = 1

예시 2)
a = 1
def vartest(a):
    a = + 1
    return a   <추가된 아웃풋>

a = vartest(a)
print(a)

<결과>
a = 2
