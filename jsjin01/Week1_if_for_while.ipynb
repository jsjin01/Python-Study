## 제어 & 반복 & 함수

# 함수
- Python의 함수는 코드의 재사용성을 높이고, 모듈화된 작업을 수행하는 데 사용됨
- def로 정의 
- 매개 변수를 받을 수 있음
- default 매개 변수를 선언 가능 
- 가변 매개 변수 *args 를 사용하여 여러 개의 인수를 한번에 받을 수 있음(list 처럼 사용 가능)
- 가변 매개 변수 **kwargs를 사용하여 여러 개의 키워드 인수를 받을 수 있음(Dictionary 처럼 사용 가능)
- return을 이용하여 반환값 반환할 수 있음

* 전역변수와 지역 변수 
전역 변수(Global Variable): 함수 외부에서 선언되어 프로그램 전체에서 접근 가능.
지역 변수(Local Variable): 함수 내부에서 선언되어 해당 함수에서만 사용 가능.

- 함수 내부에서 전역 변수를 변경하기 위해서는 global 키워드를 선언해야 됨 

# 제어문 (if)
- python에는 Switch 문이 따로 존재하지 않음
- if() : 형태로 사용
- elif() 로 계속 이어서 사용할 수 있음 
- else 위 조건들에 해당하지 않는 모든 경우를 처리할 수 있음

# 반복문(while & for)

* while문
- while(조건) : 조건이 참인 동안 계속 반복 실행됨

* for 문 
- for 변수 in 시퀀스(리스트, 문자열 등) : 시퀀스를 순회하며 반복 실행
- for 변수 in range(start, stop, step) :
start 값만 설정 시 :  0부터 start까지
stop 까지 설정 시 : start부터 stop - 1까지
step 설정 시 : 증가 값이 설정 됨



```python
x : int = 10
i : int = 0 
def global_Change():
    global x
    x = 30

def sum_numbers(*args):
    print(args)
    return sum(args)

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")


print(f"변경 전 : x : {x}")
global_Change()
print(f"변경 후 : x : {x}\n")

print(sum_numbers(1,2,3,4,5,0.4))
print()
print_info(name = "전상진", age ="25")
print()

while(i <6):
    i += 1
    print("*"*i)
print()

for j in range(1, 10 , 2):
    print("*"*j)
print()

for k in "python":
    print(k)
```
