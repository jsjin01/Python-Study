# 자료구조

* 자료 = 데이터 ---> 저장공간 + 읽기,쓰기,삽입,삭제,탐색

    - 입력 데이터를 가지고 논리적인 문제를 해결하는 것이 알고리즘이다.
    - 알고리즘: 유한한 횟수의 연산들을 이용하여 정답을 출력하는것

* 자료구조의 예
    1. 기본 데이터 구조 : List, Tuple, Dictionary, set
    2. 고급 데이터 구조 : Stack, Queue, Linked List, Tree, Graph, Heap

# 리스트

리스트는 순서가 있는 **변경 가능한** 컬렉션으로 다양한 데이터 타입을 포함할 수 있다.

예시)
my_list = [1, 2, 3, 'hello', 4.5]   # 한 리스트에 다양한 타입 포함 가능


# 튜플

튜플은 순서가 있는 **변경 불가능한** 컬렉션으로 리스트와 비슷하지만, 튜플의 값을 변경할 수 없다는 차이점이 있다.

예시)
my_tuple = (1, 2, 3, 'hello', 4.5)


# 세트

세트는 순서가 없는 변경 가능한 컬렉션으로, 중복된 값을 허용하지 않는다. 수학의 집합 개념과 유사하다.

예시)
my_set = {1, 2, 3, 4, 5}

# 딕셔너리

딕셔너리는 키-값 쌍으로 구성된 순서가 없는 변경 가능한 컬렉션이다. 키는 고유한 값을 가져야 하고, 값은 중복된 값을 허용한다.

예시)
my_dict = {'name': 'John', 'age': 25, 'city': 'Seoul'}


# 스택(Stack)

* 스택은 LIFO(Last In, First Out) 구조로, **가장 나중에** 삽입된 요소가 **가장 먼저** 삭제되는 구조이다. 파이썬에서는 리스트를 이용하여 스택을 구현할 수 있다.

* 스택은 주로 두 가지 기본 연산을 지원한다.
    1. 푸쉬(push): 스택의 맨 위에 요소를 추가한다.
    2. 팝(pop): 스택의 맨 위에 있는 요소를 제거하고 반환한다.
* 이 외에 스택의 상태를 확인하기 위해 추가될 수 있는 연산
    1. 피크(peak): 스택의 맨 위에 있는 요소를 제거하지 않고 반환한다.
    2. 비어있는지 확인(isEmpty): 스택이 비어있는지 확인한다.

# 큐(Queue)

* 큐는 FIFO(Fisrst In, First Out) 구조로, **가장 먼저** 삽입된 요소가 가장 먼저 삭제되는 구조이다. 파이썬의 collections.deque를 사용하여 구현할 수 있다.(또는 queue 모듈을 사용)

* 일반적으로 두 가지 주요 작업을 수행한다.
    - enqueue: 큐의 끝에 데이터를 추가한다.
    - dequeue: 큐의 앞에서 데이터를 제거한다.

# 링크드 리스트(Linked List)

* 링크드 리스트는 연결된 노드들의 시퀀스를 나타내는 자료 구조이다. 각 노드는 데이터를 저장하고, 다음 노드를 가리키는 포인터(참조)를 포함한다. 배열과는 달리 **동적인 크기**를 가지며, 삽입 및 삭제가 배열보다 효율적일 수 있다.
    * 배열이 비효율적인 이유
    
    - 배열은 고정된 크기로 크기를 늘리거나 줄이는 작업을 할 때 새로운 배열을 생성하고 데이터를 복사해야 한다.
    - 배열의 중간에 요소를 삽입하거나 삭제할 떄, 나머지 요소들을 이동해야 한다. 이는 시간 복잡도가 O(n)으로, 데이터의 양이 많아질수록 비효율적이다.

* 링크드 리스트에는 단일 연결 리스트와 이중 연결 리스트가 있다.

# 트리(Tree)

* 트리는 계층적 구조를 가지며, 루트 노드에서 시작하여 자식 노드로 확장된다.
    * 이진 트리
        - 이진 트리는 각 노드가 최대 두 개의 자식 노드를 가질 수 있는 트리이다.
        - 이진 트리는 다음과 같은 구조를 가진다.
            * 루트 노드: 트리의 최상위 노드이다.
            * 내부 노드: 적어도 하나의 자식 노드를 가진 노드이다.
            * 단말 노드(리프 노드): 자식 노드가 없는 노드이다.
    
    * 이진 탐색 트리
        - 이진 탐색 트리는 이진 트리의 일종으로, 각 노드의 왼쪽 서브트리에는 해당 노드보다 작은 값들이, 오른쪽 서브트리에는 큰 값들이 저장된다. 이를 통해 효율적인 검색, 삽입, 삭제 작업이 가능하다ㅏ.
        - 이진 탐색 트리의 주요 특성:
            * 왼쪽 서브트리의 모든 노드 값은 루트 노드보다 작다.
            * 오른쪽 서브트리의 모든 노드 값은 루트 노드보다 크다.
            * 각 서브트리 역시 이진 탐색 트리이다.
        

# 그래프(Graph)

* 그래프는 정점(Vertex)와 간선(Edge)으로 구성된 자료 구조로, 객체들 간의 관계를 표현하는 데 유용하다. 그래프는 방향 그래프와 무방향 그래프로 나눌 수 있다.

    * 정점: 그래프에서 개별 객체를 나타내는 것이다. 정점은 다양한 것을 표현할 수 있으며, 각 정점은 고유한 이름이나 값을 가질 수 있다.
        - 예를 들어 소셜 네트워크 그래프에서는 각 정점이 사용자를 나타낼 수 있다. 도시 간의 도로망 그래프에서는 각 정점이 도시를 나타낼 수 있다.
    
    * 간선: 간선은 두 정점을 연결하는 선이다. 간선은 정점 간의 관계나 경로를 나타낸다.
        - 예를 들어 소셜 네트워크 그래프에서는 간선이 두 사용자 간의 친구 관계를 나타낼 수 있다. 도시 간의 도로망 그래프에서는 간선이 두 도시 간의 도로를 나타낼 수 있다.

    * 방향 간선과 무방향 간선
        * 방향 간선: 방향성을 가지는 간선으로, 한 정점에서 다른 정점으로의 단방향 경로를 나타낸다.
        * 무방향 간선: 방향성을 가지지 않는 간선으로, 두 정점 사이의 쌍방향 경로를 나타낸다.
    

* 그래프의 표현
    - 그래프를 표현하는 방법은 여러 가지가 있지만, 주로 두 가지를 많이 사용한다.
        1. 인접 리스트: 각 정점에 연결된 다른 정점들의 리스트를 저장한다.
        2. 인접 행렬: 정점들 간의 연결 여부를 행렬로 나타낸다.

#인접 리스트를 이용한 그래프 표현
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],

각 정점이 키(key)이고, 연결된 정점들의 리스트가 값(Value)인 딕셔너리로 그래프 표현

    [1, 0, 0, 0, 0, 1],  # C

2차원 리스트(행렬)로 그래프 표현한 것으로, 행과 열의 인덱스는 정점을 나타내며, 값은 간선의 존재 여부를 나타낸다.(0은 간선이 없음을, 1은 간선이 있음을 의미)

# 그래프 탐색

* 그래프 탐색은 주어진 그래프에서 특정 정점을 찾거나, 그래프의 모든 정점을 방문하는 과정을 의미한다.
    - 깊이 우선 탐색(DFS): 재귀나 스택을 이용하여 그래프를 깊게 탐색하는 방법이다.
    - 너비 우선 탐색(BFS): 너비 우선 탐색은 큐를 이용하여 그래프를 넓게 탐색하는 방법이다.

# 힙(Heap)

* 힙은 완전 이진 트리 기반의 자료 구조로, 항상 가장 작은 또는 가장 큰 요소를 빠르게 찾을 수 있게 해준다. 힙은 주로 우선순위 큐(Priority Queue)와 같은 곳에서 사용된다. 파이썬은 heapq 모듈을 이용하여 힙을 쉽게 구현할 수 있다. 

* 최소 힙(Min-Heap)
    * 최소 힙은 부모 노드의 값이 자식 노드의 값보다 항상 작거나 같은 구조를 가진다. 항상 최솟값을 빠르게 찾을 수 있다.
        * 주요 함수
            - heappush(heap, item): 힙에 요소를 추가한다.
            - heappop(heap): 힙에서 최솟값을 제거하고 반환한다.
            - heapify(iterable): 주어진 반복 가능한 객체를 힙으로 변환한다.

* 최대 힙(Max-Heap)
    * 최대 힙은 부모 노드의 값이 자식 노드의 값보다 항상 크거나 같은 구조를 가진다. 항상 최댓값을 빠르게 찾을 수 있다. heapq 모듈은 기본적으로 최소 힙을 제공하기 때문에 약간의 변형이 필요하다.

* 힙의 주요 기능
    1. 삽입: 새로운 요소를 힙에 추가할 때, 힙의 속성을 유지하면서 삽입된다. O(log n)의 시간 복잡도를 가진다.
    2. 삭제: 힙에서 최솟값(또는 최솟값)을 제거할 때, 힙의 속성을 유지하면서 제거된다. O(log n)의 시간 복잡도를 가진다.
    3. 힙 생성: 주어진 리스트를 힙 구조로 변환한다. O(n)의 시간 복잡도를 가진다.


```python
# 스택 예시
stack = []
# 요소 삽입
stack.append(1)
stack.append(2)
stack.append(3)
print(stack)  # [1, 2, 3]

# 요소 삭제
print(stack.pop())  # 3
print(stack.pop())  # 2
print(stack)  # [1]

# 스택 구현하기
#스택 클래스 정의
class Stack:    
    #__init__메서드는 클래스가 인스턴스화 될 때 호출되는 생성자
    def __init__(self):
        #빈 리스트로 초기화, 스택의 요소들을 저장하는 데 사용 
        self.stack = []     
    
    #push 메서드 정의. 리스트의 append를 사용하여 item을 스택에 추가
    def push(self, item):
        self.stack.append(item)
    
    #pop 메서드는 스택의 맨 위에 있는 요소를 제거하고 반환
    def pop(self):     
        #스택이 비어있는지 확인 후, pop 메서드를 사용하여 요소를 제거 
        if not self.is_empty():     
            return self.stack.pop()
        else:
            return "Stack is empty"
    
    #스택의 맨 위에 있는 요소를 제거하지 않고 반환
    def peek(self):     
        #스택이 비어있지 않은 경우, 마지막 요소를 반환
        if not self.is_empty():     
            return self.stack[-1]
        #스택이 비어있는 경우 메시지 반환
        else:                       
            return "Stack is empty"
    
    #스택이 비어있는지 확인
    def is_empty(self):     
        #스택의 길이가 0이면 True, 그렇지 않으면 False 반환
        return len(self.stack) == 0     
    
    #스택의 크기 반환
    def size(self):     
        return len(self.stack)

# 스택 사용 예제
s = Stack()
s.push(1)
s.push(2)
s.push(3)
print("Stack after pushes:", s.stack)  # Stack after pushes: [1, 2, 3]
print("Popped item:", s.pop())  # Popped item: 3
print("Stack after pop:", s.stack)  # Stack after pop: [1, 2]
print("Peek item:", s.peek())  # Peek item: 2
print("Stack size:", s.size())  # Stack size: 2
```

```python
# 큐 예시
from collections import deque

#빈 deque 객체를 생성하고, queue 변수에 저장
queue = deque()

# 요소 삽입
queue.append(1)
queue.append(2)
queue.append(3)
print(queue)  # deque([1, 2, 3])

# 요소 삭제
print(queue.popleft())  # 1
print(queue.popleft())  # 2
print(queue)  # deque([3])
```

```python
#단일 연결 리스트 예시
class Node:     #노드를 나타내는 클래스 정의 
    def __init__(self, data):
        self.data = data  # 노드의 데이터
        self.next = None  # 다음 노드를 가리키는 포인터 초기화

#단일 연결 리스트를 나타내는 클래스 정의
class SinglyLinkedList:
    #단일 연결 리스트 클래스의 생성자
    def __init__(self):
        self.head = None  # 리스트의 머리를 초기화

    #리스트의 끝에 데이터를 추가
    def append(self, data):
        new_node = Node(data)
        if not self.head:   #리스트가 비어있는 경우, 새로운 노드를 머리로 설정
            self.head = new_node
            return
        last = self.head    #리스트의 마지막 노드를 찾기 위해 초기화
        #마지막 노드를 찾을 때까지 반복
        while last.next:
            last = last.next    #다음 노드로 이동
        last.next = new_node    #마지막 노드의 다음 포인터를 새로운 노드로 설정

    def print_list(self):
        current = self.head #현재 노드를 머리로 초기화
        while current:  #리스트의 모든 노드를 순회
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# 단일 연결 리스트 사용 예제
sll = SinglyLinkedList()
sll.append(1)
sll.append(2)
sll.append(3)
print("단일 연결 리스트")
sll.print_list()  # 1 -> 2 -> 3 -> None


#이중 연결 리스트 예제
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None    #이전 노드를 가리키는 포인터 초기화

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last    #새로운 노드의 이전 포인터를 마지막 노드로 설정 

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print("None")

# 이중 연결 리스트 사용 예제
dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.append(3)
print("이중 연결 리스트")
dll.print_list()  # 1 <-> 2 <-> 3 <-> None


#링크드 리스트의 삽입 예제
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    #주어진 노드 뒤에 새로운 노드를 삽입하는 메서드 
    def insert_after(self, prev_node, data):
        if not prev_node:
            print("이전 노드를 찾을 수 없습니다.")
            return
        new_node = Node(data)
        new_node.next = prev_node.next  #새로운 노드의 다음 포인터를 이전 노드의 다음 노드로 설정(새로운 노드의 다음 포인터 설정)
        prev_node.next = new_node   #이전 노드의 다음 포인터를 새로운 노드로 설정(새로운 노드의 이전 포인터 설정)

    def print_list(self):
        current = self.head #현재 노드를 머리로 초기화
        while current:  #리스트의 모든 노드를 순회
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# 예제 실행
sll = SinglyLinkedList()
sll.head = Node(1)
second = Node(2)
third = Node(3)
sll.head.next = second
second.next = third
print("삽입 전 리스트")
sll.print_list()    
sll.insert_after(second, 4)  # 노드 2 뒤에 노드 4 삽입
print("삽입 후 리스트")
sll.print_list()

#링크드 리스트의 삭제 예제
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # 주어진 값을 가진 노드를 삭제하는 메서드
    def delete_node(self, key):
        temp = self.head    #temp 변수를 리스트의 머리로 초기화
        if temp is not None:
            if temp.data == key:    #temp 노드의 데이터가 key와 일치하는지 확인 
                self.head = temp.next   #리스트의 머리를 다음 노드로 설정 
                temp = None     #temp를 None으로 설정하여 노드 삭제
                return
        while temp is not None:     #temp가 None이 아닌 경우 반복
            if temp.data == key:
                break
            prev = temp     #prev 변수를 현재 노드로 설정 
            temp = temp.next    #temp를 다음 노드로 이동
        if temp == None:    #temp가 None인 경우 메서드 종료
            return
        prev.next = temp.next   #이전 노드의 다음 포인터를 temp의 다음 노드로 설정
        temp = None

    def print_list(self):
        current = self.head #현재 노드를 머리로 초기화
        while current:  #리스트의 모든 노드를 순회
            print(current.data, end=" -> ")
            current = current.next
        print("None")


# 예제 실행
sll = SinglyLinkedList()
sll.head = Node(1)
second = Node(2)
third = Node(3)
sll.head.next = second
second.next = third
print("삭제 전 리스트")
sll.print_list()
sll.delete_node(2)  # 노드 2 삭제
print("삭제 후 리스트")
sll.print_list()
```

```python
# 이진 트리의 기본 구현
#노드를 나타내는 클래스 정의 
class Node:
    def __init__(self, key):
        self.left = None    #노드의 왼쪽 자식 초기화
        self.right = None   #노드의 오른쪽 자식 초기화
        self.val = key  #노드의 값 초기화

#중위 순회를 수행하는 함수 
def inorder(root):
    if root:    #노드가 존재하는지 확인
        inorder(root.left)  #왼쪽 서브트리를 중위 순회
        print(root.val, end=" ")    #현재 노드의 값 출력
        inorder(root.right) #오른쪽 서브트리 중위 순회
'''
중위 순회란 왼쪽 서브트리, 현재 노드를 방문, 오른쪽 서브트리를 중위 순회하는 방식이다.
이진 탐색 트리의 경우 노드 값이 오름차순으로 정렬된 리스트를 얻을 수 있다.
'''

#트리에 노드를 삽입하는 함수
def insert(root, key):
    if root is None:    #현재 노드가 None인 경우 새로운 노드를 생성하여 반환
        return Node(key)
    else:   #현재 노드가 None이 아닌 경우, 삽입할 위치를 찾는다.
        if root.val < key:  #현재 노드의 값이 삽입할 값보다 작을 경우 오른쪽 서브트리에 삽입
            root.right = insert(root.right, key)
        else:   #현재 노드의 값이 삽입할 값보다 크거나 같을 경우 왼쪽 서브트리에 삽입
            root.left = insert(root.left, key)
    return root    #현재 노드 반환

# 이진 트리 생성 및 예제 실행
r = Node(50)
r = insert(r, 30)
r = insert(r, 20)
r = insert(r, 40)
r = insert(r, 70)
r = insert(r, 60)
r = insert(r, 80)

print("중위 순회 결과:")
inorder(r)  # 20 30 40 50 60 70 80


#이진 탐색 트리 예제
'''
#삽입 위와 동일
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root
'''

#삭제
#노드를 삭제하는 메서드 
def deleteNode(root, key):
    if root is None:    #현재 노드가 None인 경우, 그대로 반환 
        return root

    if key < root.val:  #삭제할 값이 현재 노드보다 작은 경우, 왼쪽 서브트리에서 삭제를 계속한다.
        root.left = deleteNode(root.left, key)
    elif key > root.val:    #삭제할 값이 현재 노드보다 큰 경우, 오른쪽 서브트리에서 삭제를 계속한다.
        root.right = deleteNode(root.right, key)
    else:   #삭제할 값이 현재 노드의 값과 일치하는 경우
        if root.left is None:   #왼쪽 자식이 없는 경우, 오른쪽 자식을 반환
            return root.right
        elif root.right is None:    #오른쪽 자식이 없는 경우, 왼쪽 자식을 반환
            return root.left
        
        temp = minValueNode(root.right) #오른쪽 서브트리에서 가장 작은 값을 찾는다.
        root.val = temp.val    #현재 노드의 값을 가장 작은 값으로 대체한다.
        root.right = deleteNode(root.right, temp.val)   #오른쪽 서브트리에서 가장 작은 값을 삭제한다.

    return root

#가장 작은 값을 찾는 메서드 
def minValueNode(node):
    current = node  #현재 노드 초기화 
    while current.left is not None: #왼쪽 자식이 존재하는 동안 반복
        current = current.left
    return current


#노드 삭제 
r = deleteNode(r, 20) 
print("중위 순회 결과 (노드 20 삭제 후):") 
inorder(r) # 30 40 50 60 70 80 
print("\n") 
r = deleteNode(r, 30) 
print("중위 순회 결과 (노드 30 삭제 후):") 
inorder(r) # 40 50 60 70 80 
print("\n") 
r = deleteNode(r, 50) 
print("중위 순회 결과 (노드 50 삭제 후):") 
inorder(r) # 40 60 70 80 
print("\n")
```

```python
# 그래프 예시
class Graph:
    def __init__(self):
        self.graph = {}
    
    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)
    
    def print_graph(self):
        for node in self.graph:
            print(node, '->', ' -> '.join(map(str, self.graph[node])))

g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)
g.print_graph()
# 1 -> 2 -> 3
# 2 -> 4
# 3 -> 5

# 인접 리스트를 이용한 그래프 표현
graph = {
    'A': ['B', 'C'],    #정점 'A'는 정점 'B'와 'C'에 연결되어 있다.
    'B': ['A', 'D', 'E'],   #정점 'B'는 정점 'A','D','E'에 연결되어 있다.
    'C': ['A', 'F'],    #정점 'C'는 정점 'A'와 'F'에 연결되어 있다.
    'D': ['B'],    #정점 'D'는 정점 'B'에 연결되어 있다.
    'E': ['B', 'F'],    #정점 'E'는 정점 'B'와 'F'에 연결되어 있다.
    'F': ['C', 'E']    #정점 'F'는 정점 'C'와 'E'에 연결되어 있다.
}

# 인접 행렬을 이용한 그래프 표현
graph = [
    [0, 1, 1, 0, 0, 0],  # A
    [1, 0, 0, 1, 1, 0],  # B
    [1, 0, 0, 0, 0, 1],  # C
    [0, 1, 0, 0, 0, 0],  # D
    [0, 1, 0, 0, 0, 1],  # E
    [0, 0, 1, 0, 1, 0]   # F
]
'''
graph[0]은 정점 A와 관련된 간선 정보를 나타낸다.
    graph[0][1] == 1: 정점 A와 정점 B가 연결되어 있다.
    graph[0][2] == 1: 정점 A와 정점 C가 연결되어 있다.
graph[1]은 정점 B와 관련된 간선 정보를 나타낸다.
    graph[1][0] == 1: 정점 B와 정점 A가 연결되어 있다.
    graph[1][3] == 1: 정점 B와 정점 D가 연결되어 있다.
    .
    .
    .

그래프 구조 시각화:
  A -- B     C
  |    |     |
  |    |     |
  C    D     F
       |     |
       E ----

'''


#깊이 우선 탐색(DFS)
#DFS 함수 정의, graph는 그래프를 나타내고, start는 시작 정점이다. visited는 방문한 정점들을 저장하는 집합이다.
def dfs(graph, start, visited=None):
    if visited is None:    #visited가 None이라면 빈 집합으로 초기화
        visited = set()    
    visited.add(start)    #현재 정점을 방문했다고 표시
    print(start, end=" ")   #현재 정점을 출력
    
    #현재 정점과 연결된 모든 정점에 대해 반복
    for next in graph[start]:
        if next not in visited:
            dfs(graph, next, visited)
    return visited

# 예제 실행
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
print("DFS 방문 순서:")
dfs(graph, 'A')
print("\n")

#너비 우선 탐색(BFS)
# deque를 사용하기 위해 collections 모듈 가져오기 
from collections import deque

#BFS 함수 정의. graph는 그래프, start는 시작 정점 
def bfs(graph, start):
    visited = set() #방문한 정점들을 저장하는 빈 집합을 초기화 
    queue = deque([start])  #시작 정점을 큐에 추가
    visited.add(start)  #시작 정점을 방문했다고 표시

    #큐가 비어 있지 않은 동안 반복
    while queue:
        vertex = queue.popleft()    #큐에서 정점을 하나 꺼낸다.
        print(vertex, end=" ")    #현재 정점을 출력한다.

        for next in graph[vertex]:
            if next not in visited: #정점이 아직 방문되지 않은 경우
                visited.add(next)   #정점을 방문했다고 표시
                queue.append(next)  #정점을 큐에 추가

# 예제 실행
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
print("BFS 방문 순서:")
bfs(graph, 'A')

```

```python
# 힙 예시
import heapq

heap = []
# 요소 삽입
heapq.heappush(heap, 3)
heapq.heappush(heap, 1)
heapq.heappush(heap, 2)
print(heap)  # [1, 3, 2]

# 요소 삭제
print(heapq.heappop(heap))  # 1
print(heap)  # [2, 3]


#최소 힙
# 힙 생성
heap = []

# 요소 추가
heapq.heappush(heap, 10)
heapq.heappush(heap, 20)
heapq.heappush(heap, 5)

print("힙:", heap)  # [5, 20, 10]. 힙은 항상 최소 요소를 최상위에 둔다.

# 최솟값 제거 및 반환
min_item = heapq.heappop(heap)
print("제거된 최솟값:", min_item)  # 5
print("힙:", heap)  # [10, 20]


#최대 힙
# 힙 생성
max_heap = []

# 요소 추가 (최대 힙을 구현하기 위해 음수로 변환)
heapq.heappush(max_heap, -10)
heapq.heappush(max_heap, -20)
heapq.heappush(max_heap, -5)

print("힙 (음수로 변환):", max_heap)  # [-20, -10, -5]

# 최댓값 제거 및 반환 (음수로 변환 후)
max_item = -heapq.heappop(max_heap)
print("제거된 최댓값:", max_item)  # 20
print("힙 (양수로 변환):", [-x for x in max_heap])  # [10, 5]
```
