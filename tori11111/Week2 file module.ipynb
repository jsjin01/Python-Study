###### 파일

파일 생성하기
# newfile.py
f = open("새파일,txt", 'w')
f.close()

파일 열기 모드
r : 읽기 모드 - 파일을 읽기만 할 때 사용
w : 쓰기 모드 - 파일에 내용을 쓸 때 사용
a : 추가 모드 - 파일의 마지막에 새로운 내용을 추가할 때 사용한다.

파일을 쓰기 모드(r)로 열면 해당 파일이 이미 존재할 경우 원래 있는 내용이 전부 사라지고 없다면 새로운 파일이 생성됨.

새파일.txt를 C:/Python_Study 디렉토리에 생성하고 싶을 때
# newfile2.py
f = open("C:/Python_Study/새파일,txt", 'w')
f.close()

f.close()는 열려 있는 파일 객체를 닫아주는 역할을 함

파일 경로와 슬래시
1. /로 사용
2. \\로 사용
3. r"C:/Python_Study/새파일,txt"

파일을 쓰기 모드로 열어 내용 쓰기
1번 방식
# write_data.py
f = open("C:/doit/새파일.txt", 'w')
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()

2번 방식
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    print(data)

1번 방식은 파일에 데이터를 적는 방법, 2번 방식 모니터 화면에 데이터를 출력하는 방법이다.

#####모듈

모듈이란 - 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일
다른 파이썬 프로그램에서 불러와 사용할 수 있도록 만들어 둔 파이썬 파일

계산기 모듈
# mod1.py
def add(a,b)
    return a + b

def sub(a,b)
    return a - b

import 모듈_이름 - 이미 만들어 놓은 파이썬 모듈을 사용하게 해줌

함수 이름 하나만 쓰는 경우
from 모듈_이름 import 모듈_함수

모듈의 이름을 붙이지 않고 사용
from 모듈_이름 import add, sub
from 모듈_이름 import *

from mod1 import add
add(3, 4)

# mod1.py
def add(a, b): 
    return a+b

def sub(a, b): 
    return a-b

if __name__ == "__main__":
    print(add(1, 4))
    print(sub(4, 2))

이럴 경우 5 2로 결과 값이 나오는 데 if __name__ == "__main__":를 쓴 이유는 모듈을 불러 사용하는 경우에는 거짓이 되어 조건문 다음 문장이 실행되지 않는다..
